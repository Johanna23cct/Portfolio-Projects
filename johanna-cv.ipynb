```python
import matplotlib.pyplot as plt

def generate_cv():
    plt.figure(figsize=(8.5, 11))
    plt.text(0.5, 0.5, "Your CV Content", ha='center', va='center', fontsize=12)
    plt.axis('off')
    plt.savefig('cv.png', dpi=300, bbox_inches='tight')

if __name__ == "__main__":
    generate_cv()
```
